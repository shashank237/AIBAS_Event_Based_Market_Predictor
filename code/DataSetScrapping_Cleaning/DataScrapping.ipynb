```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import io
import os

# --------------------------------------------------
# CONFIG
# --------------------------------------------------
URL = "https://raw.githubusercontent.com/arunaravikasturi/Potsdam-University/refs/heads/main/Readme.md"
OUTPUT_PATH = "../../data/Market_Trend_External.csv"

# Ensure output directory exists
os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)

# --------------------------------------------------
# STEP 1: FETCH RAW CONTENT
# --------------------------------------------------
response = requests.get(URL)
response.raise_for_status()

# --------------------------------------------------
# STEP 2: PARSE WITH BEAUTIFULSOUP
# (Raw GitHub content is plain text, not HTML)
# --------------------------------------------------
soup = BeautifulSoup(response.text, "html.parser")
raw_text = soup.get_text().strip()

# --------------------------------------------------
# STEP 3: CONVERT TEXT TO DATAFRAME
# --------------------------------------------------
csv_buffer = io.StringIO(raw_text)
df = pd.read_csv(csv_buffer)

# --------------------------------------------------
# STEP 4: SAVE TO CSV
# --------------------------------------------------
df.to_csv(OUTPUT_PATH, index=False)

print("‚úÖ Data extracted successfully using BeautifulSoup")
print(f"üìÅ File saved at: {OUTPUT_PATH}")
print(f"üìä Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```
